```python
import scipy.io as sio
import numpy as np
from projection import load_cameras

def compare_params(dict1, dict2):
    # Load the camera parameters from the two .mat files
    param1 = load_cameras(dict1)
    param2 = load_cameras(dict2)
    
    # Get the common cameras from both sets (assuming the same camera names are present)
    common_cameras = set(param1.keys()).intersection(set(param2.keys()))
    
    for cam in common_cameras:
        print(f"\nComparing camera: {cam}")

        # List of parameters to compare
        params_to_compare = ["K", "r", "t", "RDistort", "TDistort"]
        
        for param in params_to_compare:
            # Extract and compare each parameter
            param_val1 = np.squeeze(param1[cam][param])
            param_val2 = np.squeeze(param2[cam][param])

            # Check if the values are different
            if not np.allclose(param_val1, param_val2):
                print(f"  Difference in {param}:")
                print(f"    File 1 value:\n{param_val1}")
                print(f"    File 2 value:\n{param_val2}")
            else:
                print(f"  {param} is identical between the files.")





dir1 = '/home/lq53/mir_data/chris_new_method_test/chris_new_label3d_dannce.mat'
# dir2 = '/home/lq53/mir_data/chris_new_method_test/chrris_prev_label3d_dannce.mat' prev before update
dir2 = '/home/lq53/mir_data/chris_new_method_test/updated/chrris_prev_label3d_dannce.mat'

compare_params(dir1, dir2)
```
