```python
from pathlib import Path

# Set your base directory here
base_dir = Path("/data/big_rim/uploade_ssh_mir_dataset/oct3v1_social_mini/social")

# Find all subdirectories without metadata/frame_mappings.json
missing = [
    str(subdir) 
    for subdir in base_dir.iterdir() 
    if subdir.is_dir() and not (subdir / "metadata" / "frame_mapping.json").exists()

]

# Display results
for path in missing:
    print(path)
    
print(f"\nFound {len(missing)} subdirectories without frame_mappings.json")
```
