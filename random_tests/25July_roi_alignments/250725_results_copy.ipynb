```python
import os
import shutil

# ─── PARAMETERS ───
BASE_SRC        = '/data/big_rim/rsync_dcc_sum/Oct3V1'
BASE_DST        = '/data/big_rim/rsync_dcc_sum/Oct3V1_MIR_Aligned'
MIR_FOLDER_NAME = 'MIR_Aligned'
TXT_EXT         = '.txt'

# ─── TOGGLE DRY RUN ───
dry_run = True   # True → only print; False → perform copy
# ───────────────────────

def copy_mir_aligned(base_src, base_dst, mir_folder, txt_ext, dry_run=True):
    for root, dirs, files in os.walk(base_src):
        if mir_folder in dirs:
            # build corresponding destination path
            rel = os.path.relpath(root, base_src)
            dst_base = os.path.join(base_dst, rel)

            # plan to copy the MIR_Aligned folder
            src_mir = os.path.join(root, mir_folder)
            dst_mir = os.path.join(dst_base, mir_folder)
            print(f"[MIR_FOLDER] {src_mir} → {dst_mir}")
            if not dry_run:
                os.makedirs(dst_base, exist_ok=True)
                shutil.copytree(src_mir, dst_mir, dirs_exist_ok=True)

            # plan to copy any .txt in the same folder
            for fname in files:
                if fname.lower().endswith(txt_ext):
                    src_txt = os.path.join(root, fname)
                    dst_txt = os.path.join(dst_base, fname)
                    print(f"[   TXT    ] {src_txt} → {dst_txt}")
                    if not dry_run:
                        os.makedirs(dst_base, exist_ok=True)
                        shutil.copy2(src_txt, dst_txt)

    if dry_run:
        print("\nDry run complete. No files were copied.")
    else:
        print("\nCopy complete.")

# execute
copy_mir_aligned(BASE_SRC, BASE_DST, MIR_FOLDER_NAME, TXT_EXT, dry_run)
```
