```python
import os
import re
import json
import datetime
import pyarrow.compute as pc
from datetime import date


# ----- Helper Functions -----

def parse_time_str(time_str):
    """
    Parse a time string in the format 'HH_MM_SS' or 'HH_MM' into a tuple (hour, minute, second).
    """
    parts = time_str.split('_')
    if len(parts) == 3:
        return int(parts[0]), int(parts[1]), int(parts[2])
    elif len(parts) == 2:
        return int(parts[0]), int(parts[1]), 0
    else:
        raise ValueError(f"Invalid time format: {time_str}")

def time_to_seconds(hour, minute, second):
    return hour * 3600 + minute * 60 + second

def get_rec_time_from_videos(rec_base, date_folder, rec_file):
    """
    Given the rec_base, date_folder, and rec_file, build the videos folder path:
        videos_path = f"{rec_base}/{date_folder}/{rec_file}/videos"
    and return the time-of-day (hour, minute, second) based on its modification time.
    """
    videos_path = os.path.join(rec_base, date_folder, rec_file, "videos")
    try:
        timestamp = os.path.getmtime(videos_path)
        dt = datetime.datetime.fromtimestamp(timestamp)
        return dt.hour, dt.minute, dt.second
    except Exception as e:
        print(f"Error getting time for {videos_path}: {e}")
        return None

def build_mini_path(mini_root, animal_id, custom_label, date_str, time_str):
    """
    Build the miniscope path using:
      mini_root / animal_id / custom_label / date_str / time_str / My_V4_Miniscope
    """
    return os.path.join(mini_root, animal_id, custom_label, date_str, time_str, "My_V4_Miniscope")

def build_rec_path(rec_base, date_folder, rec_file):
    """
    Build the rec path as:
      rec_base / date_folder / rec_file / DANNCE / predict00 / save_data_AVG.mat
    """
    return os.path.join(rec_base, date_folder, rec_file, "DANNCE", "predict00", "save_data_AVG.mat")

# ----- Matching Function Using PyArrow Filtering (trimmed paths) -----

def match_mini_to_rec_all(filtered_table_mini, all_rec_table, filtered_rec_table,
                          mini_root, rec_base, time_threshold=3600):
    """
    For each processed miniscope record in filtered_table_mini, find the best matching rec candidate
    from all_rec_table by modification time and only keep the parent folders in the mapping keys/values.
    """
    mapping = {}
    
    for mini_rec in filtered_table_mini.to_pylist():
        mini_date      = mini_rec["date"]       # e.g., "2024_10_25"
        mini_time_str  = mini_rec["time"]       # e.g., "15_50_39"
        animal_id      = mini_rec["animal_id"]
        custom_label   = mini_rec["custom_label"]
        
        try:
            h, m, s = parse_time_str(mini_time_str)
        except ValueError as e:
            print(f"Skipping mini record with invalid time '{mini_time_str}': {e}")
            continue
        mini_seconds = time_to_seconds(h, m, s)
        
        # full path then trimmed to parent folder
        full_mini_path = build_mini_path(mini_root, animal_id, custom_label, mini_date, mini_time_str)
        mini_folder    = os.path.dirname(full_mini_path)
        
        # filter all_rec_table to same date
        cond_date    = pc.equal(all_rec_table.column("date_folder"), mini_date)
        rec_date_tbl = all_rec_table.filter(cond_date)
        rec_candidates = rec_date_tbl.to_pylist()
        
        best_diff = None
        best_rec  = None
        
        for rec in rec_candidates:
            rec_file = rec["rec_file"]
            rec_time = get_rec_time_from_videos(rec_base, mini_date, rec_file)
            if rec_time is None:
                continue
            rec_secs = time_to_seconds(*rec_time)
            diff = abs(mini_seconds - rec_secs)
            if best_diff is None or diff < best_diff:
                best_diff = diff
                best_rec  = rec
        
        if best_rec is not None:
            raw_rec_path = build_rec_path(rec_base, mini_date, best_rec["rec_file"])
            # go up: save_data_AVG.mat  predict00  DANNCE  rec_file folder
            rec_folder = os.path.dirname(os.path.dirname(os.path.dirname(raw_rec_path)))
            # check if this rec_file is in filtered_rec_table
            cond_filt = pc.equal(filtered_rec_table.column("rec_file"), best_rec["rec_file"])
            is_filtered = filtered_rec_table.filter(cond_filt).num_rows > 0
            valid = (best_diff is not None and best_diff <= time_threshold) and is_filtered
        else:
            rec_folder = None
            best_diff  = None
            valid      = False
        
        mapping[mini_folder] = {
            "rec_path": rec_folder,
            "time_diff": best_diff,
            "valid": valid
        }
        print(f"Mini folder: {mini_folder}\nMatched rec folder: {rec_folder} "
              f"(time diff: {best_diff}) valid: {valid}\n")
    
    return mapping

def resolve_duplicates(mapping):
    """
    Ensure each rec_path only maps to one mini_folder. If a rec_folder appears multiple times,
    keep the mini_folder with the smallest time_diff and invalidate the others.
    """
    rec_to_mini = {}
    for mini_folder, info in mapping.items():
        rec = info["rec_path"]
        if rec is None:
            continue
        rec_to_mini.setdefault(rec, []).append((mini_folder, info["time_diff"]))
    
    for rec, mini_list in rec_to_mini.items():
        if len(mini_list) > 1:
            best_mini, best_diff = min(mini_list, key=lambda x: x[1])
            for mini_folder, diff in mini_list:
                if mini_folder != best_mini:
                    mapping[mini_folder]["rec_path"] = None
                    mapping[mini_folder]["valid"]    = False
                    print(f"Duplicate resolved: {mini_folder} lost rec_path {rec} "
                          f"(time_diff {diff} > {best_diff})")
    return mapping


# these tables should already exist in your environment:
#   filtered_table_mini, all_df (all_rec_table), filtered_table (filtered_rec_table)
mini_root = "/data/big_rim/rsync_dcc_sum/Oct3V1mini_sorted"
rec_base  = "/data/big_rim/rsync_dcc_sum/Oct3V1"
save_path = os.path.join(rec_base, f"{date.today()}_mini_to_rec_mapping_trimmed.json")

mapping = match_mini_to_rec_all(
    filtered_table_mini,
    all_df,
    filtered_table,
    mini_root,
    rec_base,
    time_threshold=20
)
mapping = resolve_duplicates(mapping)

# save to JSON
with open(save_path, "w") as f:
    json.dump(mapping, f, indent=4)
print("Trimmed mapping saved to mini_to_rec_mapping_trimmed_pre_processed.json")
```

```python
#!/usr/bin/env python3
import json
from pathlib import Path

#  CONFIG 
MAPPING_FILE   = save_path  # 
TIME_THRESHOLD = 15          # seconds

#  LOAD & FILTER 
with open(MAPPING_FILE, "r") as f:
    mapping = json.load(f)

for mini_path, info in mapping.items():
    td = info.get("time_diff")
    rec = info.get("rec_path")
    if rec and td is not None and td < TIME_THRESHOLD:
        mini_dir = Path(mini_path)
        rec_dir  = Path(rec)

        # ensure folders exist
        mini_dir.mkdir(parents=True, exist_ok=True)
        rec_dir.mkdir(parents=True, exist_ok=True)

        # prepare sidecar paths
        mini_sync = mini_dir / "sync_to_rec_path.txt"
        rec_sync  = rec_dir  / "sync_to_mini_path.txt"

        # write minirec sync file if missing
        if mini_sync.exists():
            print(f"{mini_sync} already exists, skipping")
        else:
            mini_sync.write_text(rec)
            print(f"{mini_sync} wrote.")

        # write recmini sync file if missing
        if rec_sync.exists():
            print(f"{rec_sync} already exists, skipping")
        else:
            rec_sync.write_text(mini_path)
            print(f"{rec_sync} wrote.")

print(f"Done writing sidecar files for all pairs with time_diff < {TIME_THRESHOLD}s.")
```
