```python
import pandas as pd




from utlis.sync_utlis.dappy_align_utlis import load_new_filtered_data_from_h5, export_aligned_data_from_paths

# Notebook or at the top of dappy_align_utlis.py
import sys, os, glob



# now you can safely:
#   from utlis.sync_utlis.dappy_align_utlis import export_aligned_data_from_paths


# from utlis.sync_utlis.dappy_align_utlis import batch_export_and_csv # load_new_filtered_data_from_h5, export_aligned_data_from_paths

pickle_path = "/home/lq53/mir_repos/dappy_24_nov/byws_version/250116_wav_ffix_ang_pos/50_p/datastruct.p"



# def batch_export_and_csv(
#     post_ana, 
#     base_folder, 
#     pickle_path
# ):
#     """
#     For each row in post_ana.to_pandas(), build:
#       rec_path = base_folder / date_folder / rec_file
#     then:
#       - ensure rec_path/MIR_Aligned exists
#       - call export_aligned_data_from_paths(pickle_path, rec_path, frame_map, out_h5)
#       - load back the filtered_data group into a DataFrame
#       - save that DataFrame to rec_path/MIR_Aligned/filtered_data.csv
#     """
#     # 1. turn post_ana into pandas and build rec paths
#     df = post_ana.to_pandas()
#     df['rec_path'] = df.apply(
#         lambda row: os.path.join(base_folder, row['date_folder'], row['rec_file']),
#         axis=1
#     )
    
#     for _, row in df.iterrows():
#         rec_path = row['rec_path']
#         mir_dir = os.path.join(rec_path, "MIR_Aligned")
#         os.makedirs(mir_dir, exist_ok=True)
        
#         # 2. frame mapping JSON is always in MIR_Aligned/frame_mapping.json
#         frame_map = os.path.join(mir_dir, "frame_mapping.json")
        
#         # 3. name the HDF5 and CSV outputs
#         out_h5  = os.path.join(mir_dir, "filtered_data.h5")
#         out_csv = os.path.join(mir_dir, "filtered_data.csv")
        
#         print(f"\n--- Processing {rec_path} ---")
#         # 4. export the HDF5
#         export_aligned_data_from_paths(
#             pickle_path,
#             rec_path,
#             frame_map,
#             out_h5
#         )
        
#         # 5. load it back in and dump CSV
#         try:
#             df_filt = load_new_filtered_data_from_h5(out_h5)
#             df_filt.to_csv(out_csv, index=False)
#             print(f" → CSV saved to: {out_csv}")
#         except Exception as e:
#             print(f" ⚠️  Failed to load/write CSV for {rec_path}: {e}")



def batch_export_and_csv(
    post_ana, 
    base_folder, 
    pickle_path,
    dry_run=False,
):
    """
    Same as before, but when dry_run=True it only logs actions, without
    creating directories, writing HDF5s or CSVs.
    """

    OTHER_ENV = "/home/lq53/miniconda3/envs/neuroposelib"
    lib_dir   = os.path.join(OTHER_ENV, "lib")

    # find any pythonX.Y/site-packages under that env
    candidates = glob.glob(os.path.join(lib_dir, "python*/site-packages"))
    if not candidates:
        raise RuntimeError(f"No site-packages found in {lib_dir}")
    # pick the first one (or choose the one that matches your current version)
    site_pkgs = candidates[0]
    print(f"→ adding {site_pkgs} to sys.path")
    sys.path.insert(0, site_pkgs)


    df = post_ana.to_pandas()
    df['rec_path'] = df.apply(
        lambda row: os.path.join(base_folder, row['date_folder'], row['rec_file']),
        axis=1
    )
    
    for _, row in df.iterrows():
        rec_path = row['rec_path']
        mir_dir  = os.path.join(rec_path, "MIR_Aligned")
        frame_map = os.path.join(mir_dir, "frame_mapping.json")
        out_h5   = os.path.join(mir_dir, "filtered_data.h5")
        out_csv  = os.path.join(mir_dir, "filtered_data.csv")
        
        print(f"\n--- Processing {rec_path} ---")
        if dry_run:
            print(f"[DRY RUN] would mkdir    {mir_dir}")
            print(f"[DRY RUN] would export → {out_h5}")
            print(f"[DRY RUN] would load & csv → {out_csv}")
            continue
        
        os.makedirs(mir_dir, exist_ok=True)
        export_aligned_data_from_paths(
            pickle_path,
            rec_path,
            frame_map,
            out_h5
        )
        try:
            df_filt = load_new_filtered_data_from_h5(out_h5)
            df_filt.to_csv(out_csv, index=False)
            print(f" → CSV saved to: {out_csv}")
        except Exception as e:
            print(f" ⚠️  Failed for {rec_path}: {e}")


batch_export_and_csv(post_ana, base_folder, pickle_path) # dry_run=True
```
