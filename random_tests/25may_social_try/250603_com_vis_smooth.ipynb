```python
import os
import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt
import pandas as pd

def plot_raw_com_with_manual_intervals(
    base_folder,
    com_folder_name='COM/predict00',
    manual_ranges_s=None,
    fps=30,
    save_csv=True
):
    """
     COM mat  (x,y,z) 
    
     baseline

    
    ----
    base_folder : str
       "{com_folder_name}/com3d0.mat"
    com_folder_name : str, default 'COM/predict00'
       base_folder  COM 
    manual_ranges_s : List[Tuple[float,float]]
      
        [(0.0, 40.0), (41.0, 54.0), (54.0, 88.0), ...]
    fps : int, default 30
      /COM 
    save_csv : bool, default True
       CSV"{base_folder}/{com_folder_name}/vis/com_manual_summary.csv"

    
    ----
    pd.DataFrame
       baseline 
      - label: "social_i"  "baseline"
      - start_s, end_s (baseline  NaN)
      - start_frame, end_frame
      - mean_dist, median_dist, min_dist, max_dist, std_dist, n_frames
    """
    # 1.  com3d0.mat  reshape
    com_mat = os.path.join(base_folder, com_folder_name, 'com3d0.mat')
    if not os.path.isfile(com_mat):
        raise FileNotFoundError(f" COM : {com_mat}")

    mat = sio.loadmat(com_mat)
    if 'com' not in mat:
        raise KeyError(f"'com'  {com_mat} ")
    raw = mat['com']
    # reshape to (n_frames, 3, n_animals)
    if raw.ndim == 3 and raw.shape[1] == 3:
        com_data = raw
    elif raw.ndim == 2 and raw.shape[1] % 3 == 0:
        n_animals = raw.shape[1] // 3
        com_data = raw.reshape(-1, 3, n_animals)
    else:
        raise ValueError(f"Unexpected COM shape {raw.shape}")
    n_frames, _, n_animals = com_data.shape
    if n_animals < 2:
        raise ValueError("")

    # 2. 
    a0 = com_data[:, :, 0]  # (n_frames,3)
    a1 = com_data[:, :, 1]
    dists = np.linalg.norm(a0 - a1, axis=1)  # (n_frames,)

    # 3.  fps  ()  
    manual_ranges_s = manual_ranges_s or []
    manual_ranges_frames = []
    for (start_s, end_s) in manual_ranges_s:
        start_f = int(round(start_s * fps))
        end_f   = int(round(end_s   * fps))
        #  [0, n_frames]
        start_f = max(0, min(start_f, n_frames))
        end_f   = max(0, min(end_f,   n_frames))
        if start_f >= end_f:
            continue
        manual_ranges_frames.append((start_s, end_s, start_f, end_f))

    # 4.  + 
    frames = np.arange(n_frames)
    plt.figure(figsize=(12, 4))
    plt.plot(frames, dists, color='tab:blue', label='Raw COM Distance')
    for idx, (start_s, end_s, start_f, end_f) in enumerate(manual_ranges_frames):
        # 
        label = 'Social' if idx == 0 else None
        plt.axvspan(start_f, end_f, color='tab:gray', alpha=0.3, label=label)
    plt.xlabel("Frame Index")
    plt.ylabel("Distance (mm)")
    plt.title("Raw COM Distance with Manual Social Intervals Highlighted")
    plt.legend()
    plt.tight_layout()
    plt.show()

    # 5.  baseline
    mask_all = np.zeros(n_frames, dtype=bool)
    summary_list = []

    # 
    for idx, (start_s, end_s, start_f, end_f) in enumerate(manual_ranges_frames):
        seg = dists[start_f:end_f]
        mask_all[start_f:end_f] = True
        summary_list.append({
            'label':       f"social_{idx+1}",
            'start_s':     start_s,
            'end_s':       end_s,
            'start_frame': start_f,
            'end_frame':   end_f,
            'mean_dist':   np.nanmean(seg),
            'median_dist': np.nanmedian(seg),
            'min_dist':    np.nanmin(seg),
            'max_dist':    np.nanmax(seg),
            'std_dist':    np.nanstd(seg),
            'n_frames':    seg.size
        })

    # baseline 
    base_seg = dists[~mask_all]
    summary_list.append({
        'label':       'baseline',
        'start_s':     np.nan,
        'end_s':       np.nan,
        'start_frame': np.nan,
        'end_frame':   np.nan,
        'mean_dist':   np.nanmean(base_seg) if base_seg.size else np.nan,
        'median_dist': np.nanmedian(base_seg) if base_seg.size else np.nan,
        'min_dist':    np.nanmin(base_seg) if base_seg.size else np.nan,
        'max_dist':    np.nanmax(base_seg) if base_seg.size else np.nan,
        'std_dist':    np.nanstd(base_seg) if base_seg.size else np.nan,
        'n_frames':    base_seg.size
    })

    df_summary = pd.DataFrame(summary_list)
    print("\nManual intervals vs baseline distance summary:\n")
    print(df_summary.to_string(index=False))

    # 6.  CSV
    if save_csv:
        vis_dir = os.path.join(base_folder, com_folder_name, 'vis')
        os.makedirs(vis_dir, exist_ok=True)
        csv_path = os.path.join(vis_dir, 'com_manual_summary.csv')
        df_summary.to_csv(csv_path, index=False)
        print(f"\n{csv_path}")

    return df_summary

# ==========  ==========
if __name__ == '__main__':
    session_folder = "/data/big_rim/rsync_dcc_sum/Oct3V1/2024_10_31/2social_mini_20240819V1r1_femalebleach_11_48/"  # 
    manual_ranges_s = [
        (0.0, 40.0),   (41.0, 54.0),  (54.0, 88.0),
        (89.0, 120.0), (137.0, 149.0), (162.0, 165.0),
        (200.0, 218.0), (230.0, 243.0), (337.0, 343.0),
        (371.0, 375.0), (466.0, 468.0), (468.0, 472.0),
        (530.0, 530.0), (533.0, 540.0), (562.0, 572.0),
        (573.0, 576.0), (615.0, 617.0), (670.0, 671.0),
    ]
    df_summary = plot_raw_com_with_manual_intervals(
        base_folder=session_folder,
        com_folder_name='COM/predict00',
        manual_ranges_s=manual_ranges_s,
        fps=30,
        # save_csv=True
    )
```

```python
import os
import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.mixture import GaussianMixture

def detect_auto_social_intervals(
    dists_sm,
    rel_speed,
    fps,
    min_duration_s=1.0,
    speed_percentile=90,
    d_thresh_manual=None
):
    """
     dists_sm  rel_speed
      1.  d_thresh_manual 2  GMM  d_thresh
      2.  rel_speed  speed_percentile 
      3.  (dists_sm <= d_thresh) AND (rel_speed <= speed_thresh) 
      4.  True  min_duration_s 
    :
      auto_intervals_s:  [(start_s, end_s), ...]
      d_thresh        : 
      speed_thresh    : 
    """
    # ---- 1.  ----
    if d_thresh_manual is not None:
        d_thresh = float(d_thresh_manual)
    else:
        X_gmm = dists_sm.reshape(-1, 1)
        gmm = GaussianMixture(n_components=2, covariance_type='full', random_state=0)
        gmm.fit(X_gmm)
        mus = gmm.means_.flatten()
        sigmas = np.sqrt(gmm.covariances_).flatten()
        pis = gmm.weights_.flatten()

        mu1, mu2 = mus
        s1, s2 = sigmas
        pi1, pi2 = pis

        A = 1.0 / (2 * s1 * s1) - 1.0 / (2 * s2 * s2)
        B = mu2 / (s2 * s2) - mu1 / (s1 * s1)
        C = (mu1 * mu1) / (2 * s1 * s1) - (mu2 * mu2) / (2 * s2 * s2) + np.log((pi2 * s1) / (pi1 * s2))

        disc = B * B - 4 * A * C
        if disc < 0:
            d_thresh = float((mu1 + mu2) / 2.0)
        else:
            r1 = (-B + np.sqrt(disc)) / (2 * A)
            r2 = (-B - np.sqrt(disc)) / (2 * A)
            candidates = [r for r in (r1, r2) if min(mu1, mu2) < r < max(mu1, mu2)]
            d_thresh = float(candidates[0]) if candidates else float((mu1 + mu2) / 2.0)

    # ---- 2.  ----
    speed_thresh = np.percentile(rel_speed, speed_percentile)

    # ---- 3.  ----
    mask_dist = (dists_sm <= d_thresh)
    mask_speed = (rel_speed <= speed_thresh)
    mask_auto = mask_dist & mask_speed

    # ---- 4.  True  ----
    n_frames = len(dists_sm)
    min_frames = max(1, int(round(min_duration_s * fps)))

    auto_intervals = []
    is_active = False
    start_idx = 0
    count = 0
    for i, flag in enumerate(mask_auto):
        if flag:
            if not is_active:
                is_active = True
                start_idx = i
                count = 1
            else:
                count += 1
        else:
            if is_active:
                if count >= min_frames:
                    end_idx = i
                    auto_intervals.append((start_idx, end_idx))
                is_active = False
    if is_active and count >= min_frames:
        auto_intervals.append((start_idx, n_frames))

    auto_intervals_s = [(s / fps, e / fps) for (s, e) in auto_intervals]
    return auto_intervals_s, d_thresh, speed_thresh


def plot_com_with_manual_and_auto(
    base_folder,
    com_folder_name='COM/predict00',
    manual_segments=None,
    fps=30,
    min_duration_s=1.0,
    smoothing_window_s=0.1,
    speed_percentile=90,
    d_thresh_manual=None
):
    """
    
      1.  COM 
      2. 
      3. 
      4. 

    :
      base_folder        :  COM/predict00/com3d0.mat
      com_folder_name    : COM 
      manual_segments    :  [(label, start_s, end_s), ...]
      fps                : 
      min_duration_s     : 
      smoothing_window_s : 
      speed_percentile   : 
      d_thresh_manual    :  None GMM 
    :
      df_summary         :  + baseline  DataFrame
      auto_intervals_s   :  [(start_s, end_s), ...]
    """
    # ---- 1.  COM  ----
    com_mat = os.path.join(base_folder, com_folder_name, 'com3d0.mat')
    if not os.path.isfile(com_mat):
        raise FileNotFoundError(f" COM : {com_mat}")

    mat = sio.loadmat(com_mat)
    if 'com' not in mat:
        raise KeyError(f"'com'  {com_mat} ")

    raw = mat['com']
    # reshape to (n_frames, 3, n_animals)
    if raw.ndim == 3 and raw.shape[1] == 3:
        com_data = raw
    elif raw.ndim == 2 and raw.shape[1] % 3 == 0:
        n_animals = raw.shape[1] // 3
        com_data = raw.reshape(-1, 3, n_animals)
    else:
        raise ValueError(f"Unexpected COM shape {raw.shape}")

    n_frames, _, n_animals = com_data.shape
    if n_animals < 2:
        raise ValueError("")

    # ---- 2.  ----
    a0 = com_data[:, :, 0]
    a1 = com_data[:, :, 1]
    dists = np.linalg.norm(a0 - a1, axis=1)  # shape: (n_frames,)

    # ---- 3.  ----
    window_frames = max(1, int(round(smoothing_window_s * fps)))
    kernel = np.ones(window_frames) / window_frames
    dists_sm = np.convolve(dists, kernel, mode='same')

    # ---- 4.  rel_speed ----
    dd = np.diff(dists_sm, prepend=dists_sm[0])
    pos1 = a0
    pos2 = a1
    v1 = np.linalg.norm(np.diff(pos1, axis=0, prepend=pos1[:1, :]), axis=1) * fps
    v2 = np.linalg.norm(np.diff(pos2, axis=0, prepend=pos2[:1, :]), axis=1) * fps
    rel_speed = np.abs(dd) * fps

    # ---- 5.  ----
    auto_intervals_s, d_thresh, speed_thresh = detect_auto_social_intervals(
        dists_sm=dists_sm,
        rel_speed=rel_speed,
        fps=fps,
        min_duration_s=min_duration_s,
        speed_percentile=speed_percentile,
        d_thresh_manual=d_thresh_manual
    )

    # ---- 6.  ----
    manual_ranges_frames = []
    if manual_segments:
        for (_, start_s, end_s) in manual_segments:
            start_f = max(0, min(int(round(start_s * fps)), n_frames))
            end_f   = max(0, min(int(round(end_s * fps)), n_frames))
            if start_f < end_f:
                manual_ranges_frames.append((start_s, end_s, start_f, end_f))

    # ---- 7.  summary DataFrame ----
    mask_social = np.zeros(n_frames, dtype=bool)
    summary_list = []

    # 
    for idx, (s0, s1) in enumerate(auto_intervals_s):
        start_f = int(round(s0 * fps))
        end_f   = int(round(s1 * fps))
        seg = dists_sm[start_f:end_f]
        mask_social[start_f:end_f] = True
        summary_list.append({
            'label':       f"auto_social_{idx+1}",
            'start_s':     s0,
            'end_s':       s1,
            'start_frame': start_f,
            'end_frame':   end_f,
            'mean_dist':   np.nanmean(seg),
            'median_dist': np.nanmedian(seg),
            'min_dist':    np.nanmin(seg),
            'max_dist':    np.nanmax(seg),
            'std_dist':    np.nanstd(seg),
            'n_frames':    seg.size
        })

    # baseline 
    base_seg = dists_sm[~mask_social]
    summary_list.append({
        'label':       'baseline',
        'start_s':     np.nan,
        'end_s':       np.nan,
        'start_frame': np.nan,
        'end_frame':   np.nan,
        'mean_dist':   np.nanmean(base_seg) if base_seg.size else np.nan,
        'median_dist': np.nanmedian(base_seg) if base_seg.size else np.nan,
        'min_dist':    np.nanmin(base_seg) if base_seg.size else np.nan,
        'max_dist':    np.nanmax(base_seg) if base_seg.size else np.nan,
        'std_dist':    np.nanstd(base_seg) if base_seg.size else np.nan,
        'n_frames':    base_seg.size
    })

    df_summary = pd.DataFrame(summary_list)

    # ---- 8.  &  ----
    frames = np.arange(n_frames)
    plt.figure(figsize=(12, 4))
    plt.plot(frames, dists_sm, color='tab:blue', label='Smoothed COM Distance')

    # : 
    for idx, (start_s, end_s, start_f, end_f) in enumerate(manual_ranges_frames):
        label = 'Manual Social' if idx == 0 else None
        plt.axvspan(start_f, end_f, color='tab:gray', alpha=0.3, label=label)

    # : 
    for idx, (s0, s1) in enumerate(auto_intervals_s):
        start_f = int(round(s0 * fps))
        end_f   = int(round(s1 * fps))
        label = 'Auto Social' if idx == 0 else None
        plt.axvspan(start_f, end_f, color='tab:red', alpha=0.2, label=label)

    # 
    plt.axhline(d_thresh, color='red', linestyle='--', label=f'Distance Threshold={d_thresh:.2f}')

    plt.xlabel("Frame Index")
    plt.ylabel("Distance (same units as COM)")
    plt.title("Smoothed COM Distance with Manual & Autodetected Social Intervals")
    plt.legend(loc='upper right')
    plt.tight_layout()
    plt.show()

    # ---- 9.  +  ----
    plt.figure(figsize=(8, 4))
    plt.hist(dists_sm, bins=50, alpha=0.6, density=True, label='All Distances')
    plt.axvline(d_thresh, color='red', linestyle='--', label='Distance Threshold')
    plt.xlabel("Smoothed Distance")
    plt.ylabel("Density")
    plt.title("Distance Distribution & Threshold")
    plt.legend()
    plt.tight_layout()
    plt.show()

    # ---- 10.  vs   ----
    print("\n  Social  () ")
    if manual_segments:
        for (label, s0, s1) in manual_segments:
            print(f"  {label}: [{s0:.2f}, {s1:.2f}]")
    else:
        print("  ()")

    print("\n  Social  () ")
    for (s0, s1) in auto_intervals_s:
        print(f"  [{s0:.2f}, {s1:.2f}]")

    return df_summary, auto_intervals_s


# ======  ======
if __name__ == '__main__':
    # ==== 1.  ====
    session_folder = "/data/big_rim/rsync_dcc_sum/Oct3V1/2024_10_31/2social_mini_20240819V1r1_single_11_29/"

    # ==== 2. Social: (label, start_s, end_s) ====
    manual_segments = [
        ('sniff',          74.0,  79.0),    # 01:1401:19
        ('follow_partner', 79.0,  81.0),    # 01:1901:21
        ('sniff_around',   81.0,  92.0),    # 01:2101:32
        ('follow',         92.0,  94.0),    # 01:3201:34
        ('parallel',       94.0,  97.0),    # 01:3401:37
        ('follow_tail',    97.0, 110.0),    # 01:3701:50
        ('sniff_around',  131.0, 133.0),    # 02:1102:13
        ('follow_tail',   133.0, 138.0),    # 02:1302:18
        ('sniff',         138.0, 139.0),    # 02:1802:19
    ]

    # ==== 3.  ====
    #  d_thresh_manual=239  239 
    df_summary, auto_segs = plot_com_with_manual_and_auto(
        base_folder=session_folder,
        com_folder_name='COM/predict00',
        manual_segments=manual_segments,
        fps=30,
        min_duration_s=1.0,
        smoothing_window_s=0.1,
        speed_percentile=90,
        d_thresh_manual=239
    )

    # ==== 4.  CSV ====
    vis_dir = os.path.join(session_folder, 'COM', 'predict00', 'vis')
    os.makedirs(vis_dir, exist_ok=True)
    csv_path = os.path.join(vis_dir, 'com_auto_vs_manual_summary.csv')
    df_summary.to_csv(csv_path, index=False)
    print(f"\n{csv_path}")
```

```python
import numpy as np

def interval_clip(interval, cutoff):
    """
     (start_s, end_s)  [0, cutoff] 
     None (new_start, new_end)
    """
    start, end = interval
    new_start = max(0.0, start)
    new_end   = min(cutoff, end)
    if new_start >= new_end:
        return None
    return (new_start, new_end)

def intervals_overlap(intv1, intv2):
    """
     (start_s, end_s)  > 0
    """
    a0, a1 = intv1
    b0, b1 = intv2
    return not (a1 <= b0 or b1 <= a0)

def overlap_duration(intv1, intv2):
    """
     (start_s, end_s)  0
    """
    if not intervals_overlap(intv1, intv2):
        return 0.0
    return max(0.0, min(intv1[1], intv2[1]) - max(intv1[0], intv2[0]))


if __name__ == '__main__':
    #  manual_segments  auto_segs 
    #  (label, start_s, end_s) (start_s, end_s) 
    manual_segments = [
        ('sniff',          74.0,  79.0),
        ('follow_partner', 79.0,  81.0),
        ('sniff_around',   81.0,  92.0),
        ('follow',         92.0,  94.0),
        ('parallel',       94.0,  97.0),
        ('follow_tail',    97.0, 110.0),
        ('sniff_around',  131.0, 133.0),
        ('follow_tail',   133.0, 138.0),
        ('sniff',         138.0, 139.0),
    ]
    #  start/end 
    manual_intervals = [(s, e) for (_, s, e) in manual_segments]

    #  auto_segs  detect  [(start_s, end_s), ...]
    #  auto_segs 
    auto_segs = [
        # 
        (74.03, 77.83),
(78.77, 80.33),
(80.37, 81.43),
(82.50, 83.53),
(83.57, 88.20),
(89.10, 90.10),
(90.27, 92.73),
(95.33, 97.70),
(100.17, 101.37),
(108.30, 111.37),
(129.50, 132.30),
(132.37, 133.50),
(133.63, 135.40),
(135.57, 138.10),


    ]

    # 1.  2:19 = 139 
    cutoff = 2 * 60 + 19  # 139 

    # 2.  [0, cutoff] 
    manual_cut = []
    for intv in manual_intervals:
        clipped = interval_clip(intv, cutoff)
        if clipped is not None:
            manual_cut.append(clipped)

    auto_cut = []
    for intv in auto_segs:
        clipped = interval_clip(intv, cutoff)
        if clipped is not None:
            auto_cut.append(clipped)

    # 3. 
    manual_hit = []
    for m in manual_cut:
        hit = any(intervals_overlap(m, a) for a in auto_cut)
        manual_hit.append((m, hit))

    # 4. 
    auto_hit = []
    for a in auto_cut:
        hit = any(intervals_overlap(a, m) for m in manual_cut)
        auto_hit.append((a, hit))

    # 5. 
    print("===  0139s  &  ===")
    for (start, end), hit in manual_hit:
        print(f"  [{start:.2f}, {end:.2f}]  => {'' if hit else ''}")

    print("\n===  0139s  &  ===")
    for (start, end), hit in auto_hit:
        print(f"  [{start:.2f}, {end:.2f}]  => {'' if hit else ''}")

    # 6. Precision/Recall
    #   -  0139s 
    #   -  0139s 
    #   - 
    total_manual_dur = sum(e - s for s, e in manual_cut)
    total_auto_dur   = sum(e - s for s, e in auto_cut)

    #  (m, a) 
    #  0.01s  0139s 
    ts = np.arange(0.0, cutoff, 0.01)
    is_manual = np.zeros_like(ts, dtype=bool)
    is_auto   = np.zeros_like(ts, dtype=bool)

    # 
    for s, e in manual_cut:
        mask = (ts >= s) & (ts < e)
        is_manual[mask] = True
    # 
    for s, e in auto_cut:
        mask = (ts >= s) & (ts < e)
        is_auto[mask] = True

    overlap_mask = is_manual & is_auto
    total_overlap_dur = overlap_mask.sum() * 0.01  # 0.01s 

    recall = total_overlap_dur / total_manual_dur if total_manual_dur > 0 else 0.0
    precision = total_overlap_dur / total_auto_dur if total_auto_dur > 0 else 0.0

    print(f"\n (0139s): {total_manual_dur:.2f} ")
    print(f" (0139s): {total_auto_dur:.2f} ")
    print(f": {total_overlap_dur:.2f} ")
    print(f" Recall = / = {recall:.3f}")
    print(f" Precision = / = {precision:.3f}")
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np

# ==== 1. label, start_s, end_s ====
manual_segments = [
    ('sniff',          74.0,  79.0),    # 01:1401:19
    ('follow_partner', 79.0,  81.0),    # 01:1901:21
    ('sniff_around',   81.0,  92.0),    # 01:2101:32
    ('follow',         92.0,  94.0),    # 01:3201:34
    ('parallel',       94.0,  97.0),    # 01:3401:37
    ('follow_tail',    97.0, 110.0),    # 01:3701:50
    ('sniff_around',  131.0, 133.0),    # 02:1102:13
    ('follow_tail',   133.0, 138.0),    # 02:1302:18
    ('sniff',         138.0, 139.0),    # 02:1802:19
]
manual_intervals = [(label, s, e) for (label, s, e) in manual_segments]

# ==== 2.  (start_s, end_s) ====
#  auto_segs 
auto_segs = [
(74.03, 77.83),
(78.77, 80.33),
(80.37, 81.43),
(82.50, 83.53),
(83.57, 88.20),
(89.10, 90.10),
(90.27, 92.73),
(95.33, 97.70),
(100.17, 101.37),
(108.30, 111.37),
(129.50, 132.30),
(132.37, 133.50),
(133.63, 135.40),
(135.57, 138.10),

]

# ==== 3.  ====
cutoff = 2 * 60 + 19  # 139 

# ==== 4.  [0, cutoff] ====
def interval_clip(interval, cutoff):
    """
     (start_s, end_s)  [0, cutoff] 
     None (new_start, new_end)
    """
    start, end = interval
    new_start = max(0.0, start)
    new_end   = min(cutoff, end)
    if new_start >= new_end:
        return None
    return (new_start, new_end)

# ==== 5.  ====
def overlap_duration(intv1, intv2):
    """
     (start_s, end_s)  0
    """
    a0, a1 = intv1
    b0, b1 = intv2
    if a1 <= b0 or b1 <= a0:
        return 0.0
    return max(0.0, min(a1, b1) - max(a0, b0))

# ==== 6.  [0, cutoff] ====
manual_cut = []
for (label, s, e) in manual_intervals:
    clipped = interval_clip((s, e), cutoff)
    if clipped is not None:
        manual_cut.append((label, clipped[0], clipped[1]))

auto_cut = []
for (s, e) in auto_segs:
    clipped = interval_clip((s, e), cutoff)
    if clipped is not None:
        auto_cut.append((clipped[0], clipped[1]))

# ==== 7.  ====
auto_with_labels = []  #  (start_s, end_s, assigned_label)
for (a_start, a_end) in auto_cut:
    best_label = 'none'
    best_ov    = 0.0
    for (m_label, m_start, m_end) in manual_cut:
        ov = overlap_duration((a_start, a_end), (m_start, m_end))
        if ov > best_ov:
            best_ov    = ov
            best_label = m_label
    auto_with_labels.append((a_start, a_end, best_label))

# ==== 8.  ====
labels = set(label for (label, _, _) in manual_cut)
labels.add('none')
labels = sorted(labels)

cmap = plt.get_cmap('tab10')
label2color = {}
for idx, lab in enumerate(labels):
    label2color[lab] = cmap(idx / max(1, len(labels) - 1))

# ==== 9.  ====
fig, ax = plt.subplots(figsize=(12, 3))

# 9.1  (y = 1.0)
y_manual = 1.0
bar_height = 0.2
for (m_label, m_start, m_end) in manual_cut:
    rect = mpatches.Rectangle(
        (m_start, y_manual - bar_height/2),
        width=(m_end - m_start),
        height=bar_height,
        facecolor='lightgray',
        edgecolor='k',
        alpha=0.8
    )
    ax.add_patch(rect)
    x_center = (m_start + m_end) / 2
    ax.text(
        x_center, y_manual + bar_height/2 + 0.02,
        m_label,
        ha='center',
        va='bottom',
        fontsize=9
    )

# 9.2  (y = 0.5)
y_auto = 0.5
for (a_start, a_end, a_label) in auto_with_labels:
    color = label2color[a_label]
    rect = mpatches.Rectangle(
        (a_start, y_auto - bar_height/2),
        width=(a_end - a_start),
        height=bar_height,
        facecolor=color,
        edgecolor='k',
        alpha=0.8
    )
    ax.add_patch(rect)
    x_center = (a_start + a_end) / 2
    ax.text(
        x_center, y_auto + bar_height/2 + 0.02,
        a_label,
        ha='center',
        va='bottom',
        fontsize=9
    )

# 9.3  139 
ax.axvline(cutoff, color='red', linestyle='--', linewidth=1)

# 9.4 
ax.set_ylim(0, 1.4)
ax.set_xlim(0, cutoff + 1)
ax.set_yticks([y_auto, y_manual])
ax.set_yticklabels(['Autodetected', 'Manual'], fontsize=10)
ax.set_xlabel("Time (seconds)", fontsize=11)
ax.set_title("Manual vs. Autodetected Intervals (0139s) with Labels", fontsize=12)

# 9.5 
legend_patches = []
for lab in labels:
    legend_patches.append(mpatches.Patch(color=label2color[lab], label=lab))
ax.legend(
    handles=legend_patches,
    title="Label",
    bbox_to_anchor=(1.02, 1),
    loc='upper left',
    borderaxespad=0
)

plt.tight_layout()
plt.show()
```
