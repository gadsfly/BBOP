```python
import os
import glob
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, leaves_list
from scipy.stats import zscore

# ----------------------------------------------------------------------------
# 1. 
# ----------------------------------------------------------------------------
exclude_json_path = '/home/lq53/mir_repos/BBOP/random_tests/25feb_more_corr_explo/neuro_exclude.json'
with open(exclude_json_path, 'r') as f:
    exclude_dict = json.load(f)

def get_excluded_neurons_for_session(session_path, exclude_dict):
    """
     session_path 
    """
    if session_path in exclude_dict:
        return exclude_dict[session_path]
    for key, val in exclude_dict.items():
        if key in session_path:
            return val
    return []

# ----------------------------------------------------------------------------
# 2.  session  HDF5 
# ----------------------------------------------------------------------------
def load_session_data(rec_path):
    """
     rec_path/MIR_Aligned  .h5  DataFrame 
    """
    h5_dir = os.path.join(rec_path, 'MIR_Aligned')
    pattern = '*aligned_predictions_with_ca_and_dF_F*.h5'
    h5_files = glob.glob(os.path.join(h5_dir, pattern))
    if not h5_files:
        raise FileNotFoundError(f"No .h5 with aligned_predictions_with_ca_and_dF_F in {h5_dir}")
    hdf5_file_path = h5_files[0]

    df = pd.read_hdf(hdf5_file_path, key='df')
    norm_path = os.path.normpath(rec_path)
    df['session_id']      = os.path.basename(norm_path)
    df['recording_date']  = os.path.basename(os.path.dirname(norm_path))
    df['experiment']      = os.path.basename(os.path.dirname(os.path.dirname(norm_path)))
    df['session_path']    = rec_path
    df['file_path']       = hdf5_file_path
    return df

# ----------------------------------------------------------------------------
# 3. + ROI + Z-score
# ----------------------------------------------------------------------------
def process_neuron_activity(df, exclude_dict=None, manual_exclude_indices=None, apply_exclusion=True):
    """
     dF/F  z-score 
    
      - normalized:  (n_kept_neurons, n_timepoints)  numpy  (Z-scored)
      - filtered_cols: 
      - df_reset:  DataFrame  dF/F 
    """
    df_reset = df.reset_index(drop=True)

    if apply_exclusion:
        if manual_exclude_indices is not None:
            excluded = manual_exclude_indices
        elif exclude_dict is not None:
            excluded = get_excluded_neurons_for_session(df_reset['session_path'].iloc[0], exclude_dict)
        else:
            excluded = []
    else:
        excluded = []

    exclude_cols = {f'dF_F_roi{i}' for i in excluded}
    neuron_cols = [c for c in df_reset.columns if c.startswith('dF_F_roi') and c not in exclude_cols]
    activity = df_reset[neuron_cols].values.T  # shape: (n_neurons, n_timepoints)
    variances = np.var(activity, axis=1)
    thresh = np.percentile(variances, 5)  #  5% 
    keep_mask = variances > thresh
    filtered = activity[keep_mask, :]
    filtered_cols = [col for i, col in enumerate(neuron_cols) if keep_mask[i]]
    normalized = zscore(filtered, axis=1)  #  ROI  Z-score

    return normalized, filtered_cols, df_reset

# ----------------------------------------------------------------------------
# 4. 
# ----------------------------------------------------------------------------
def find_social_close_onsets(rec_path, distance_threshold):
    """
     rec_path/MIR_Aligned  com_distances.csv com_distances_filtered.csv
     distance_threshold  > threshold  <= threshold 
     camera_frame_sixcam
    """
    aligned_dir = os.path.join(rec_path, 'MIR_Aligned')
    #  com_distances.csv com_distances_filtered.csvframe
    full_csv = os.path.join(aligned_dir, 'com_distances.csv')
    filtered_csv = os.path.join(aligned_dir, 'com_distances_filtered.csv')

    if os.path.exists(full_csv):
        df_com = pd.read_csv(full_csv)
        if 'frame' not in df_com.columns or 'distance' not in df_com.columns:
            raise KeyError("com_distances.csv  'frame'  'distance' ")
        # /
        df_com = df_com.sort_values('frame').reset_index(drop=True)
        distances = df_com['distance'].values
        frames = df_com['frame'].astype(int).values

        #  > threshold  <= threshold 
        above_mask = distances > distance_threshold
        below_mask = distances <= distance_threshold
        # shift  above below
        above_prev = np.concatenate(([False], above_mask[:-1]))
        onset_mask = np.logical_and(above_prev, below_mask)
        onset_frames = frames[onset_mask]

        return onset_frames.tolist()

    elif os.path.exists(filtered_csv):
        #  filtered CSV
        df_filt = pd.read_csv(filtered_csv)
        frame_col = df_filt.columns[0]  #  'frame'
        onset_frames = df_filt[frame_col].astype(int).values
        return onset_frames.tolist()
    else:
        raise FileNotFoundError(f" {aligned_dir}  com_distances.csv  com_distances_filtered.csv")

# ----------------------------------------------------------------------------
# 5. 
# ----------------------------------------------------------------------------
def peri_event_neuron_activity(rec_path, distance_threshold=250,
                               pre_frames=30, post_frames=60,
                               exclude_dict=None, apply_exclusion=True):
    """
     rec_path  session
      1.  HDF5 Z-score
      2. camera_frame_sixcam
      3.  df_reset  onset_frame 
      4.  pre_frames/frame  post_frames/frame 
         
      5.  shape  (n_events, n_neurons, pre_frames+post_frames+1)  3D 
      6. 
         a ROI  ROI 
         b ROI 
    
      - rec_path:  session  MIR_Aligned  HDF5
      - distance_threshold:  CSV  mm
      - pre_frames: 
      - post_frames: 
      - exclude_dict:  process_neuron_activity
      - apply_exclusion: 
    """
    #  1.  neuron activity  
    df = load_session_data(rec_path)
    neuron_norm, neuron_cols, df_reset = process_neuron_activity(
        df, exclude_dict, manual_exclude_indices=None, apply_exclusion=apply_exclusion
    )
    n_neurons, n_timepoints = neuron_norm.shape

    #  df_reset  'camera_frame_sixcam' 
    if 'camera_frame_sixcam' not in df_reset.columns:
        raise KeyError("'camera_frame_sixcam'  HDF5 ")

    #  frame -> df_reset   frame 
    frame_to_index = {}
    for idx, frame in enumerate(df_reset['camera_frame_sixcam'].values.astype(int)):
        if frame not in frame_to_index:
            frame_to_index[frame] = idx

    #  2.  onset events  
    onset_frames = find_social_close_onsets(rec_path, distance_threshold)
    if len(onset_frames) == 0:
        print(f"No social-close onsets found for session: {rec_path}")
        return

    #  3.  peri-event   
    window_length = pre_frames + post_frames + 1
    peri_data = []  #  (n_neurons, window_length) 

    for frame in onset_frames:
        if frame not in frame_to_index:
            #  frame  df_reset 
            continue
        center_idx = frame_to_index[frame]
        start_idx = center_idx - pre_frames
        end_idx = center_idx + post_frames
        #  [0, n_timepoints) 
        if start_idx < 0 or end_idx >= n_timepoints:
            # 
            continue
        #  ROI 
        snippet = neuron_norm[:, start_idx:end_idx+1]  #  end_idx
        peri_data.append(snippet)

    if len(peri_data) == 0:
        print(f"No valid peri-event windows within data range for session: {rec_path}")
        return

    peri_data = np.stack(peri_data, axis=0)  # shape: (n_events, n_neurons, window_length)
    n_events = peri_data.shape[0]

    #  4.   
    # (a)  shape  (n_neurons, window_length)
    mean_across_events = np.mean(peri_data, axis=0)

    # (b)  ROI  shape  (window_length,)
    grand_mean = np.mean(mean_across_events, axis=0)

    #  5.   
    # 5.1  ROI 
    fig1, ax1 = plt.subplots(figsize=(10, 6))
    #  ROI  mean_across_events  ROI 
    #  ROI 
    post_window = mean_across_events[:, pre_frames:pre_frames+post_frames]
    roi_var = np.var(post_window, axis=1)
    order = np.argsort(-roi_var)  # 
    clustered_data = mean_across_events[order, :]
    #  ~1/30 
    framerate = 30.0  # 
    time_axis = (np.arange(window_length) - pre_frames) / framerate

    im = ax1.pcolormesh(
        time_axis,
        np.arange(n_neurons),
        clustered_data,
        cmap='viridis',
        shading='auto'
    )
    ax1.axvline(0, color='white', linestyle='--', linewidth=1)  # 
    ax1.set_ylabel('Neuron (clustered)')
    ax1.set_xlabel('Time from event onset (s)')
    ax1.set_title('Peri-event Mean Z-scored Activity (grouped by ROI)')
    fig1.colorbar(im, ax=ax1, label='Z-score F/F')

    # 5.2  ROI 
    fig2, ax2 = plt.subplots(figsize=(8, 4))
    ax2.plot(time_axis, grand_mean, color='blue', linewidth=2)
    ax2.axvline(0, color='black', linestyle='--', linewidth=1)
    ax2.set_xlabel('Time from event onset (s)')
    ax2.set_ylabel('Mean Z-score F/F')
    ax2.set_title('Grand Average Activity Across All ROIs')
    plt.tight_layout()

    plt.show()

    #  6.  vs   
    #  ROI  pre  post  Wilcoxon 
    from scipy.stats import wilcoxon
    baseline_window = slice(0, pre_frames)                # 
    response_window = slice(pre_frames, pre_frames+post_frames)  # 
    p_values = []
    for i in range(n_neurons):
        #  ROI shape=(window_length,)
        roi_signal = mean_across_events[i, :]
        pre_vals = roi_signal[baseline_window]
        post_vals = roi_signal[response_window]
        #  pre  post 
        pre_mean = np.mean(pre_vals)
        post_mean = np.mean(post_vals)
        # 
        #  pre/post 
        #  peri_data  shape=(n_events, window_length)  pre/post  wilcoxon
        event_pre_means  = np.mean(peri_data[:, i, baseline_window], axis=1)
        event_post_means = np.mean(peri_data[:, i, response_window], axis=1)
        if len(event_pre_means) == len(event_post_means) and len(event_pre_means) >= 5:
            stat, p = wilcoxon(event_pre_means, event_post_means)
        else:
            p = np.nan
        p_values.append(p)

    #  ROIp < 0.05
    p_values = np.array(p_values)
    sig_neurons = np.where(p_values < 0.05)[0]
    print(f"Number of ROIs with significant increase after event (p < 0.05): {len(sig_neurons)}")
    #  ROI 
    sig_roi_names = [neuron_cols[i] for i in sig_neurons]
    print("Significant ROI names:", sig_roi_names)

# ----------------------------------------------------------------------------
# 7. 
# ----------------------------------------------------------------------------
if __name__ == '__main__':
    #  session_path 
    session_paths = [
    # "/data/big_rim/rsync_dcc_sum/Oct3V1/2024_12_18/20240919v1l5r1mini_p20240717PMC_social_test_11_30",
    # "/data/big_rim/rsync_dcc_sum/Oct3V1/2024_12_31/20240919v1l5r2mini_p20240717PMC_social_14_04",
    # "/data/big_rim/rsync_dcc_sum/Oct3V1/2024_10_31/2social_mini_20240819V1r1_femalebleach_11_48"


    "/data/big_rim/rsync_dcc_sum/Oct3V1/2024_12_18/20240919v1l5r1mini_p20240717PMC_social_test_11_30",
    # "/data/big_rim/rsync_dcc_sum/Oct3V1/2025_02_27/20241015PMCBE1mini_p20241015PMCRE1_12_33",
    "/data/big_rim/rsync_dcc_sum/Oct3V1/2024_12_31/20240919v1l5r2mini_p20240717PMC_social_14_04",
    # "/data/big_rim/rsync_dcc_sum/Oct3V1/2024_11_01/2social_mini_20240910V1r_AO_single_12_50",
    # "/data/big_rim/rsync_dcc_sum/Oct3V1/2024_11_01/2social_mini_20240819V1r1_AO_single_14_30",
    "/data/big_rim/rsync_dcc_sum/Oct3V1/2024_10_31/2social_mini_20240819V1r1_single_11_29",
    "/data/big_rim/rsync_dcc_sum/Oct3V1/2024_10_31/2social_mini_20240819V1r1_femalebleach_11_48",
    "/data/big_rim/rsync_dcc_sum/Oct3V1/2025_05_16/20241216V1RE1Fmini_p20241216RE2",
    "/data/big_rim/rsync_dcc_sum/Oct3V1/2025_05_16/20241216V1RE1Fmini_p20241224PMCLE1",
]


    #  250 mm
    threshold_value = 250
    #  30  1  60  2 
    pre_win  = 30
    post_win = 120

    for path in session_paths:
        try:
            print(f"\nProcessing session: {path}")
            peri_event_neuron_activity(
                rec_path=path,
                distance_threshold=threshold_value,
                pre_frames=pre_win,
                post_frames=post_win,
                exclude_dict=exclude_dict,
                apply_exclusion=True
            )
        except Exception as e:
            print(f"Failed for session {path}: {e}")
```

```python
import os
import glob
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ----------------------------------------------------------------------------
# 1. 
# ----------------------------------------------------------------------------
exclude_json_path = '/home/lq53/mir_repos/BBOP/random_tests/25feb_more_corr_explo/neuro_exclude.json'
with open(exclude_json_path, 'r') as f:
    exclude_dict = json.load(f)

def get_excluded_neurons_for_session(session_path, exclude_dict):
    if session_path in exclude_dict:
        return exclude_dict[session_path]
    for key, val in exclude_dict.items():
        if key in session_path:
            return val
    return []

# ----------------------------------------------------------------------------
# 2.  HDF5 
# ----------------------------------------------------------------------------
def load_session_data(rec_path):
    h5_dir = os.path.join(rec_path, 'MIR_Aligned')
    pattern = '*aligned_predictions_with_ca_and_dF_F*.h5'
    h5_files = glob.glob(os.path.join(h5_dir, pattern))
    if not h5_files:
        raise FileNotFoundError(f"No .h5 in {h5_dir}")
    df = pd.read_hdf(h5_files[0], key='df')
    norm_path = os.path.normpath(rec_path)
    df['session_id']     = os.path.basename(norm_path)
    df['recording_date'] = os.path.basename(os.path.dirname(norm_path))
    df['experiment']     = os.path.basename(os.path.dirname(os.path.dirname(norm_path)))
    df['session_path']   = rec_path
    df['file_path']      = h5_files[0]
    return df

# ----------------------------------------------------------------------------
# 3. 5% Z-score
# ----------------------------------------------------------------------------
def process_neuron_raw(df, exclude_dict=None, manual_exclude_indices=None, apply_exclusion=True):
    df_reset = df.reset_index(drop=True)

    if apply_exclusion:
        if manual_exclude_indices is not None:
            excluded = manual_exclude_indices
        elif exclude_dict is not None:
            excluded = get_excluded_neurons_for_session(df_reset['session_path'].iloc[0], exclude_dict)
        else:
            excluded = []
    else:
        excluded = []

    exclude_cols = {f'dF_F_roi{i}' for i in excluded}
    neuron_cols = [c for c in df_reset.columns if c.startswith('dF_F_roi') and c not in exclude_cols]
    activity = df_reset[neuron_cols].values.T  #  (n_neurons, n_timepoints)
    variances = np.var(activity, axis=1)
    thresh = np.percentile(variances, 5)  #  5% 
    keep_mask = variances > thresh

    raw_kept = activity[keep_mask, :]  #  F/F
    kept_cols = [col for i, col in enumerate(neuron_cols) if keep_mask[i]]

    return raw_kept, kept_cols, df_reset

# ----------------------------------------------------------------------------
# 4.  onset 
# ----------------------------------------------------------------------------
def find_social_close_onsets(rec_path, distance_threshold):
    aligned_dir = os.path.join(rec_path, 'MIR_Aligned')
    full_csv = os.path.join(aligned_dir, 'com_distances.csv')
    filtered_csv = os.path.join(aligned_dir, 'com_distances_filtered.csv')

    if os.path.exists(full_csv):
        df_com = pd.read_csv(full_csv).sort_values('frame').reset_index(drop=True)
        distances = df_com['distance'].values
        frames    = df_com['frame'].astype(int).values

        above_mask = distances > distance_threshold
        below_mask = distances <= distance_threshold
        above_prev = np.concatenate(([False], above_mask[:-1]))
        onset_mask = np.logical_and(above_prev, below_mask)
        onset_frames = frames[onset_mask]
        return onset_frames.tolist()

    elif os.path.exists(filtered_csv):
        df_filt = pd.read_csv(filtered_csv)
        frame_col = df_filt.columns[0]
        return df_filt[frame_col].astype(int).values.tolist()
    else:
        raise FileNotFoundError(f"Cannot find com_distances.csv or com_distances_filtered.csv in {aligned_dir}")

# ----------------------------------------------------------------------------
# 5.  F/F 
# ----------------------------------------------------------------------------
def peri_event_raw_activity(rec_path, distance_threshold=250, pre_frames=30, post_frames=60, exclude_dict=None, apply_exclusion=True):
    # 1.  F/F
    df = load_session_data(rec_path)
    raw_data, neuron_cols, df_reset = process_neuron_raw(df, exclude_dict, None, apply_exclusion)
    n_neurons, n_timepoints = raw_data.shape

    if 'camera_frame_sixcam' not in df_reset.columns:
        raise KeyError("'camera_frame_sixcam' ")

    # frame 
    frame_to_idx = {}
    for idx, frame in enumerate(df_reset['camera_frame_sixcam'].values.astype(int)):
        if frame not in frame_to_idx:
            frame_to_idx[frame] = idx

    # 2.  onset 
    onset_frames = find_social_close_onsets(rec_path, distance_threshold)
    if len(onset_frames) == 0:
        print("No onsets found.")
        return

    # 3.  peri-event 
    window_len = pre_frames + post_frames + 1
    peri_snippets = []

    for frame in onset_frames:
        if frame not in frame_to_idx:
            continue
        center_idx = frame_to_idx[frame]
        start_idx = center_idx - pre_frames
        end_idx   = center_idx + post_frames
        if start_idx < 0 or end_idx >= n_timepoints:
            continue
        snippet = raw_data[:, start_idx:end_idx+1]  # (n_neurons, window_len)
        peri_snippets.append(snippet)

    if not peri_snippets:
        print("No valid windows within data range.")
        return

    peri_array = np.stack(peri_snippets, axis=0)  # (n_events, n_neurons, window_len)
    n_events   = peri_array.shape[0]

    # 4.  F/F 
    #  n_events  -> (n_neurons, window_len)
    mean_peri = np.mean(peri_array, axis=0)
    #  n_neurons  -> (window_len,)
    grand_mean = np.mean(mean_peri, axis=0)

    # 5. 
    framerate = 30.0  # 
    time_axis = (np.arange(window_len) - pre_frames) / framerate

    fig1, ax1 = plt.subplots(figsize=(10, 6))
    #  ROI 
    post_window = mean_peri[:, pre_frames:pre_frames+post_frames]
    abs_peak = np.max(post_window, axis=1)
    order = np.argsort(-abs_peak)  #  ROI 
    data_ordered = mean_peri[order, :]
    im = ax1.pcolormesh(
        time_axis,
        np.arange(n_neurons),
        data_ordered,
        cmap='viridis',
        shading='auto'
    )
    ax1.axvline(0, color='white', linestyle='--', linewidth=1)
    ax1.set_ylabel('Neuron (sorted by abs-peak)')
    ax1.set_xlabel('Time from event onset (s)')
    ax1.set_title('Peri-event Mean F/F (raw, grouped by ROI)')
    fig1.colorbar(im, ax=ax1, label='F/F (absolute units)')

    fig2, ax2 = plt.subplots(figsize=(8, 4))
    ax2.plot(time_axis, grand_mean, linewidth=2)
    ax2.axvline(0, color='black', linestyle='--', linewidth=1)
    ax2.set_xlabel('Time from event onset (s)')
    ax2.set_ylabel('Mean F/F (absolute)')
    ax2.set_title('Grand Average F/F Across All ROIs')
    plt.tight_layout()
    plt.show()

    # 6.  0~0.5s  ROI
    response_window = np.logical_and(time_axis >= 0, time_axis <= 0.5)
    # mean_peri (n_neurons, window_len) response_window 
    peak_vals = np.max(mean_peri[:, response_window], axis=1)
    top_N = 5
    top_idxs = np.argsort(-peak_vals)[:top_N]
    print(f"Top {top_N} ROIs by absolute F/F peak (0~0.5s):")
    for idx in top_idxs:
        print(f"  ROI {neuron_cols[idx]}: peak F/F  {peak_vals[idx]:.4f}")

# ----------------------------------------------------------------------------
# 7. 
# ----------------------------------------------------------------------------
if __name__ == '__main__':
    session_paths = [
    # "/data/big_rim/rsync_dcc_sum/Oct3V1/2024_12_18/20240919v1l5r1mini_p20240717PMC_social_test_11_30",
    # "/data/big_rim/rsync_dcc_sum/Oct3V1/2024_12_31/20240919v1l5r2mini_p20240717PMC_social_14_04",
    # "/data/big_rim/rsync_dcc_sum/Oct3V1/2024_10_31/2social_mini_20240819V1r1_femalebleach_11_48"


    "/data/big_rim/rsync_dcc_sum/Oct3V1/2024_12_18/20240919v1l5r1mini_p20240717PMC_social_test_11_30",
    # "/data/big_rim/rsync_dcc_sum/Oct3V1/2025_02_27/20241015PMCBE1mini_p20241015PMCRE1_12_33",
    "/data/big_rim/rsync_dcc_sum/Oct3V1/2024_12_31/20240919v1l5r2mini_p20240717PMC_social_14_04",
    # "/data/big_rim/rsync_dcc_sum/Oct3V1/2024_11_01/2social_mini_20240910V1r_AO_single_12_50",
    # "/data/big_rim/rsync_dcc_sum/Oct3V1/2024_11_01/2social_mini_20240819V1r1_AO_single_14_30",
    "/data/big_rim/rsync_dcc_sum/Oct3V1/2024_10_31/2social_mini_20240819V1r1_single_11_29",
    "/data/big_rim/rsync_dcc_sum/Oct3V1/2024_10_31/2social_mini_20240819V1r1_femalebleach_11_48",
    "/data/big_rim/rsync_dcc_sum/Oct3V1/2025_05_16/20241216V1RE1Fmini_p20241216RE2",
    "/data/big_rim/rsync_dcc_sum/Oct3V1/2025_05_16/20241216V1RE1Fmini_p20241224PMCLE1",
]
    threshold_value = 250
    pre_win  = 30
    post_win = 60

    for path in session_paths:
        print(f"\n=== Processing {path} ===")
        try:
            peri_event_raw_activity(
                rec_path=path,
                distance_threshold=threshold_value,
                pre_frames=pre_win,
                post_frames=post_win,
                exclude_dict=exclude_dict,
                apply_exclusion=True
            )
        except Exception as e:
            print(f"Failed: {e}")
```
