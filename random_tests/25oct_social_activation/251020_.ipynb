```python
# === Solo (oct3_singles + sum24_filtered) vs Social (oct3_social) ===
import os, numpy as np, pandas as pd, matplotlib.pyplot as plt
from scipy.io import loadmat

# ---- config ----
FPS = 30.0  # speeds in mm/s; set to your true fps (None -> mm/frame)
SOCIAL_SUBDIR = "yolo_tracking/predict_train_lone_social_0829" #retrain_0917"  # your given folder
CLIP_Q = 0.999  # trim top 0.1% step spikes; keep or remove as you like

# ---- collect sessions ----
dfs = []
if 'oct3_singles'   in globals(): dfs.append(oct3_singles.to_pandas()[['rec_path']])
if 'sum24_filtered' in globals(): dfs.append(sum24_filtered.to_pandas()[['rec_path']])
solo_df = (pd.concat(dfs, ignore_index=True) if dfs else pd.DataFrame(columns=['rec_path'])).assign(condition='solo')

soc_df = pd.DataFrame(columns=['rec_path','condition'])
if 'oct3_social' in globals():
    soc_df = oct3_social.to_pandas()[['rec_path']].assign(condition='social')

meta = pd.concat([solo_df, soc_df], ignore_index=True).dropna(subset=['rec_path']).drop_duplicates('rec_path')

# ---- loaders (mm units; same .mat schema, different folders) ----
def load_xy_solo_mm(base_path):
    p1 = os.path.join(base_path, 'DANNCE', 'predict00', 'com3d_used.mat')
    p2 = os.path.join(base_path, 'COM', 'predict00', 'com3d0.mat')
    path = p1 if os.path.isfile(p1) else (p2 if os.path.isfile(p2) else None)
    if path is None: return None
    com = loadmat(path).get('com')
    if com is None or len(com) < 3: return None
    return np.asarray(com)[:, :2]  # x,y in mm


def load_xy_social_mm(base_path):
    # Check for any yolo_tracking subdirectories
    yolo_base = os.path.join(base_path, 'yolo_tracking')
    
    if not os.path.isdir(yolo_base):
        return None
    
    # Find all subdirectories with com3d0.mat
    candidates = []
    for subdir in os.listdir(yolo_base):
        mat_path = os.path.join(yolo_base, subdir, 'com3d0.mat')
        if os.path.isfile(mat_path):
            candidates.append((subdir, mat_path))
    
    if not candidates:
        return None
    
    # Sort by subdirectory name (later dates/versions come last)
    # e.g., "retrain_0829" < "retrain_0917" alphabetically
    candidates.sort(key=lambda x: x[0])
    
    # Take the latest (last in sorted order)
    _, path = candidates[-1]
    
    com = loadmat(path).get('com')
    if com is None or len(com) < 3:
        return None
    
    com = np.asarray(com)
    
    # Check if we have multiple animals (3D array)
    if com.ndim == 3:
        # Take mean position of both animals
        return np.mean(com[:, :, :2], axis=1)  # (frames, 2)
    else:
        return com[:, :2]  # (frames, 2)
    
# average, but too depenent on path... not that dynammic....
# def load_xy_social_mm(base_path):
#     path = os.path.join(base_path, SOCIAL_SUBDIR, 'com3d0.mat')
#     if not os.path.isfile(path): return None
#     com = loadmat(path).get('com')
#     if com is None or len(com) < 3: return None
#     com = np.asarray(com)
    
#     # Check if we have multiple animals (3D array)
#     if com.ndim == 3:
#         # Take mean position of both animals
#         return np.mean(com[:, :, :2], axis=1)  # (frames, 2)
#     else:
#         return com[:, :2]  # (frames, 2)

# wrong beuacse social has 2 animals idiot
# def load_xy_social_mm(base_path):
#     path = os.path.join(base_path, SOCIAL_SUBDIR, 'com3d0.mat')
#     if not os.path.isfile(path): return None
#     com = loadmat(path).get('com')
#     if com is None or len(com) < 3: return None
#     return np.asarray(com)[:, :2]  # x,y in mm

# ---- simple summary per session (no segments) ----
def summarize_total(xy, fps=FPS, clip_q=CLIP_Q):
    xy = np.asarray(xy, float)
    if xy.ndim != 2 or xy.shape[0] < 3: return None
    m = np.isfinite(xy).all(axis=1)
    xy = xy[m]
    if xy.shape[0] < 3: return None

    step = np.linalg.norm(np.diff(xy, axis=0), axis=1)  # mm/frame
    step = step[np.isfinite(step)]
    if step.size == 0: return None
    if clip_q:
        q = np.quantile(step, clip_q)
        step = step[step <= q]

    v = step * fps if (fps and fps > 0) else step
    return {
        'total_dist_mm': float(np.nansum(step)),
        'mean_speed':    float(np.nanmean(v)) if v.size else np.nan,
        'p95_speed':     float(np.nanpercentile(v, 95)) if v.size else np.nan,
    }

# ---- build df_summary ----
rows, skipped = [], []
for rec_path, cond in meta[['rec_path','condition']].itertuples(index=False):
    xy = load_xy_social_mm(rec_path) if cond == 'social' else load_xy_solo_mm(rec_path)
    if xy is None:
        skipped.append(os.path.basename(rec_path)); continue
    s = summarize_total(xy)
    if s is None:
        skipped.append(os.path.basename(rec_path)); continue
    s.update(rec_path=rec_path, condition=cond,
             session=os.path.basename(str(rec_path).rstrip(os.sep)))
    rows.append(s)

df_summary = pd.DataFrame(rows)
print(f"included: {len(df_summary)}  |  skipped: {len(skipped)}")

# ---- condition-level means ± SEM ----
grp   = df_summary.groupby('condition', sort=False)
means = grp[['total_dist_mm','mean_speed','p95_speed']].mean()
sems  = grp[['total_dist_mm','mean_speed','p95_speed']].sem(ddof=1)
print("\nCondition-level means:\n", means.round(2))

# ---- plots (three tiny bars) ----
x = np.arange(len(means.index))
labels = means.index.tolist()

plt.figure(figsize=(6,4))
plt.bar(x, means['total_dist_mm'].to_numpy(), yerr=sems['total_dist_mm'].to_numpy(), capsize=4)
plt.xticks(x, labels); plt.ylabel("Total distance (mm)"); plt.title("Total COM distance")
plt.tight_layout(); plt.show()

plt.figure(figsize=(6,4))
plt.bar(x, means['mean_speed'].to_numpy(), yerr=sems['mean_speed'].to_numpy(), capsize=4)
plt.xticks(x, labels); plt.ylabel("Mean speed (mm/s)" if (FPS and FPS>0) else "Mean speed (mm/frame)")
plt.title("Mean speed"); plt.tight_layout(); plt.show()

plt.figure(figsize=(6,4))
plt.bar(x, means['p95_speed'].to_numpy(), yerr=sems['p95_speed'].to_numpy(), capsize=4)
plt.xticks(x, labels); plt.ylabel("95th percentile speed (mm/s)" if (FPS and FPS>0) else "95th percentile speed (mm/frame)")
plt.title("95th percentile speed"); plt.tight_layout(); plt.show()
# this maybe useless becuase for some we did not filter out the conditions from summ, such as caffine... thus may need to do something different in the future....
```

```python
# Detailed debugging for skipped sessions
rows, skipped_details = [], []

for rec_path, cond in meta[['rec_path','condition']].itertuples(index=False):
    session_name = os.path.basename(str(rec_path).rstrip(os.sep))
    
    # Try loading
    xy = load_xy_social_mm(rec_path) if cond == 'social' else load_xy_solo_mm(rec_path)
    
    if xy is None:
        skipped_details.append({
            'session': session_name,
            'condition': cond,
            'reason': 'xy is None (file not found or no com data)'
        })
        continue
    
    # Try summarizing
    s = summarize_total(xy)
    
    if s is None:
        # Figure out why summarize_total failed
        xy_test = np.asarray(xy, float)
        m = np.isfinite(xy_test).all(axis=1)
        xy_clean = xy_test[m]
        
        if xy_test.ndim != 2:
            reason = f'wrong ndim: {xy_test.ndim}'
        elif xy_test.shape[0] < 3:
            reason = f'too few frames before cleaning: {xy_test.shape[0]}'
        elif xy_clean.shape[0] < 3:
            reason = f'too few frames after cleaning: {xy_clean.shape[0]} (from {xy_test.shape[0]})'
        else:
            reason = 'unknown - check step calculation'
        
        skipped_details.append({
            'session': session_name,
            'condition': cond,
            'reason': reason
        })
        continue
    
    s.update(rec_path=rec_path, condition=cond, session=session_name)
    rows.append(s)

df_summary = pd.DataFrame(rows)
df_skipped = pd.DataFrame(skipped_details)

print(f"Included: {len(df_summary)}  |  Skipped: {len(df_skipped)}")
print("\n=== Skipped sessions breakdown ===")
print(df_skipped.groupby(['condition', 'reason']).size())
print("\n=== Individual skipped sessions ===")
print(df_skipped.to_string())
```

```python
# Check what files exist in the skipped social sessions
import glob

print("=== Checking skipped social sessions ===\n")

skipped_social = df_skipped[df_skipped['condition'] == 'social']['session'].tolist()

for session in skipped_social[:5]:  # Check first 5 as examples
    # Find the full path
    matching = meta[meta['rec_path'].str.contains(session, regex=False)]
    if len(matching) == 0:
        print(f"{session}: Could not find in meta")
        continue
    
    rec_path = matching['rec_path'].iloc[0]
    print(f"\n{session}:")
    print(f"  Base path: {rec_path}")
    
    # Check if base path exists
    if not os.path.isdir(rec_path):
        print(f"  ❌ Base directory doesn't exist!")
        continue
    
    # Check for the expected file
    expected = os.path.join(rec_path, SOCIAL_SUBDIR, 'com3d0.mat')
    print(f"  Expected: {expected}")
    print(f"  Exists: {os.path.isfile(expected)}")
    
    # See what tracking directories exist
    yolo_path = os.path.join(rec_path, 'yolo_tracking')
    if os.path.isdir(yolo_path):
        subdirs = [d for d in os.listdir(yolo_path) if os.path.isdir(os.path.join(yolo_path, d))]
        print(f"  yolo_tracking subdirs: {subdirs}")
        
        # Check for com3d0.mat in any subdir
        for subdir in subdirs:
            mat_file = os.path.join(yolo_path, subdir, 'com3d0.mat')
            if os.path.isfile(mat_file):
                print(f"    ✓ Found: {subdir}/com3d0.mat")
    else:
        print(f"  ❌ No yolo_tracking directory")
    
    # Check for alternative tracking directories
    alt_dirs = glob.glob(os.path.join(rec_path, '**/com3d0.mat'), recursive=True)
    if alt_dirs:
        print(f"  Alternative com3d0.mat locations found:")
        for alt in alt_dirs:
            print(f"    - {alt.replace(rec_path, '.')}")
```
