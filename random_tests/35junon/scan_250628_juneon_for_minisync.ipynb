```python
import os
import json
import datetime
import pyarrow.compute as pc
from datetime import date

# ----- BUILD PATHS -----
def build_mini_path(mini_root, animal_id, custom_label, date_str, time_str):
    return os.path.join(
        mini_root,
        animal_id,
        custom_label,
        date_str,
        time_str,
        "My_V4_Miniscope",
        "headOrientation.csv"
    )

def build_rec_path(rec_base, date_folder, rec_file):
    return os.path.join(
        rec_base,
        date_folder,
        rec_file,
        "videos",
        "Camera1",
        "frametimes.npy"
    )

# ----- MATCH MINI TO REC USING FILE MODIFICATION TIMES -----
def match_mini_to_rec_all(filtered_table_mini,
                          all_rec_table,
                          filtered_rec_table,
                          mini_root,
                          rec_base,
                          time_threshold=3600,
                          require_prediction=True):
    mapping = {}
    for mini_rec in filtered_table_mini.to_pylist():
        date_str   = mini_rec["date"]
        time_str   = mini_rec["time"]
        animal_id  = mini_rec["animal_id"]
        custom_lbl = mini_rec["custom_label"]

        mini_path = build_mini_path(mini_root, animal_id, custom_lbl, date_str, time_str)
        try:
            mini_dt  = datetime.datetime.fromtimestamp(os.path.getmtime(mini_path))
            mini_iso = mini_dt.isoformat()
        except Exception as e:
            print(f"Skipping missing mini file: {mini_path} ({e})")
            continue

        # trimmed key: the folder just before "My_V4_Miniscope"
        trimmed_key = os.path.dirname(os.path.dirname(mini_path))

        # filter rec candidates to same date
        cond_date    = pc.equal(all_rec_table.column("date_folder"), date_str)
        rec_date_tbl = all_rec_table.filter(cond_date).to_pylist()

        best_diff    = None
        best_rec     = None
        best_rec_iso = None

        for rec in rec_date_tbl:
            rec_file = rec["rec_file"]
            rec_path = build_rec_path(rec_base, date_str, rec_file)
            if not os.path.exists(rec_path):
                continue

            rec_dt  = datetime.datetime.fromtimestamp(os.path.getmtime(rec_path))
            rec_iso = rec_dt.isoformat()
            diff    = abs((mini_dt - rec_dt).total_seconds())

            if best_diff is None or diff < best_diff:
                best_diff    = diff
                best_rec     = rec_file
                best_rec_iso = rec_iso

        if best_rec:
            rec_folder = os.path.dirname(os.path.dirname(os.path.dirname(
                build_rec_path(rec_base, date_str, best_rec)
            )))
            if require_prediction:
                cond_filt   = pc.equal(filtered_rec_table.column("rec_file"), best_rec)
                is_filtered = filtered_rec_table.filter(cond_filt).num_rows > 0
            else:
                is_filtered = True

            valid = (best_diff is not None and best_diff <= time_threshold) and is_filtered
        else:
            rec_folder   = None
            best_diff    = None
            best_rec_iso = None
            valid        = False

        mapping[trimmed_key] = {
            "mini_timestamp": mini_iso,
            "rec_timestamp":  best_rec_iso,
            "rec_path":       rec_folder,
            "time_diff_s":    best_diff,
            "valid":          valid
        }
        print(f"Mapped {trimmed_key}  {rec_folder} (={best_diff}s)")

    return mapping

# ----- DUPLICATE RESOLUTION (placeholder) -----
def resolve_duplicates(mapping):
    # insert your existing duplicate-resolution logic here
    return mapping



# ----- USAGE & SAVING -----
if __name__ == "__main__":
    mini_root = "/data/big_rim/rsync_dcc_sum/Oct3V1mini_sorted"
    rec_base  = "/data/big_rim/rsync_dcc_sum/25juneon"

    # filtered_table_mini, all_df, filtered_table must be defined / loaded beforehand
    mapping = match_mini_to_rec_all(
        filtered_table_mini,
        all_df,
        filtered_table,
        mini_root,
        rec_base,
        time_threshold=20,
        require_prediction=False
    )
    mapping = resolve_duplicates(mapping)

    save_path = os.path.join(rec_base, f"{date.today()}_mini_to_rec_mapping_juneon.json")
    with open(save_path, "w") as f:
        json.dump(mapping, f, indent=4)

    print(f"Mapping saved to {save_path}")
```

```python
#!/usr/bin/env python3
import json
from pathlib import Path

#  CONFIG 
MAPPING_FILE   = save_path  # 
TIME_THRESHOLD = 25          # seconds

#  LOAD & FILTER 
with open(MAPPING_FILE, "r") as f:
    mapping = json.load(f)

for mini_path, info in mapping.items():
    td = info.get("time_diff_s")
    rec = info.get("rec_path")
    if rec and td is not None and td < TIME_THRESHOLD:
        mini_dir = Path(mini_path)
        rec_dir  = Path(rec)

        # ensure folders exist
        mini_dir.mkdir(parents=True, exist_ok=True)
        rec_dir.mkdir(parents=True, exist_ok=True)

        # prepare sidecar paths
        mini_sync = mini_dir / "sync_to_rec_path.txt"
        rec_sync  = rec_dir  / "sync_to_mini_path.txt"

        # write minirec sync file if missing
        if mini_sync.exists():
            print(f"{mini_sync} already exists, skipping")
        else:
            mini_sync.write_text(rec)
            print(f"{mini_sync} wrote.")

        # write recmini sync file if missing
        if rec_sync.exists():
            print(f"{rec_sync} already exists, skipping")
        else:
            rec_sync.write_text(mini_path)
            print(f"{rec_sync} wrote.")

print(f"Done writing sidecar files for all pairs with time_diff < {TIME_THRESHOLD}s.")
```
