```python
import os
import re
from collections import defaultdict
from datetime import datetime

import matplotlib.pyplot as plt
import matplotlib.image as mpimg


# Convert and flatten the ChunkedArray
rec_paths = filtered_table["rec_path"].to_pylist()

base_paths = []

# Print each path
for path in rec_paths:
    # Each element is likely a list with one string, e.g., ['/some/path']
    base_paths.append(path[0] if isinstance(path, list) else path)

# 1. Collect (path, modification time) for each frametimes.mat
sessions = []
for p in base_paths:
    ft = os.path.join(p, "videos", "Camera1", "frametimes.mat")
    if os.path.isfile(ft):
        mtime = os.path.getmtime(ft)
        sessions.append((p, mtime))
    else:
        print(f"Warning: frametimes.mat not found for {p}")

# 2. Group by (day, animal)
groups = defaultdict(list)
for p, mtime in sessions:
    day = os.path.basename(os.path.dirname(p))            # e.g. "2025_06_17"
    folder = os.path.basename(p)                          # e.g. "20250407V1L23r1blkcap_econe…"
    # m = re.match(r"^(\d{8})([A-Za-z0-9]+)", folder)       # date + animalID
    # if not m:
    #     print("folder path not matched")
    #     continue
    # date_str, animal = m.groups()
    # groups[(day, animal)].append((p, mtime))
    # 2. Group by (day, animal)
    groups = defaultdict(list)
    for p, mtime in sessions:
        # same day extraction
        day = os.path.basename(os.path.dirname(p))      # e.g. "2025_06_17"
        folder = os.path.basename(p)                    # e.g. "V1L23r1_blkcap_econe…"
        # animal is everything before the first '_'
        animal = folder.split("_", 1)[0]                # e.g. "V1L23r1"
        groups[(day, animal)].append((p, mtime))







# 3. For each (day, animal), sort and plot
for (day, animal), items in groups.items():
    items.sort(key=lambda x: x[1])  # chronological
    n = len(items)
    fig, axes = plt.subplots(1, n, figsize=(n * 3, 3))
    if n == 1:
        axes = [axes]
    for ax, (p, mtime) in zip(axes, items):
        img_file = os.path.join(p, "vis", "com_circle.png")
        # if not os.path.isfilie(img_file):
        #     print(f"no image file found for {p}")
        if os.path.isfile(img_file):
            img = mpimg.imread(img_file)
            ax.imshow(img)
            ts = datetime.fromtimestamp(mtime).strftime("%H:%M:%S")
            ax.set_title(ts, fontsize=8)
        else:
            ax.text(0.5, 0.5, "no image", ha="center", va="center")
        ax.axis("off")
    fig.suptitle(f"{day}  ·  {animal}", fontsize=10)
    plt.tight_layout()
    plt.show()
```

```python
import os
from datetime import datetime

import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# 1. Flatten your rec_paths
rec_paths = filtered_table["rec_path"].to_pylist()
base_paths = [p[0] if isinstance(p, list) else p for p in rec_paths]

# 2. Collect (path, modification time) for each frametimes.mat
sessions = []
for p in base_paths:
    ft = os.path.join(p, "videos", "Camera1", "frametimes.mat")
    if os.path.isfile(ft):
        mtime = os.path.getmtime(ft)
        sessions.append((p, mtime))
    else:
        print(f"Warning: frametimes.mat not found for {p}")

# 3. Sort all sessions by timestamp
sessions.sort(key=lambda x: x[1])

# 4. Plot every session on a single figure
n = len(sessions)
fig, axes = plt.subplots(1, n, figsize=(n * 3, 3))
if n == 1:
    axes = [axes]

for ax, (p, mtime) in zip(axes, sessions):
    img_file = os.path.join(p, "vis", "com_circle.png")
    if os.path.isfile(img_file):
        img = mpimg.imread(img_file)
        ax.imshow(img)
    else:
        ax.text(0.5, 0.5, "no image", ha="center", va="center")
    ts = datetime.fromtimestamp(mtime).strftime("%Y-%m-%d %H:%M:%S")
    ax.set_title(ts, fontsize=8)
    ax.axis("off")

plt.tight_layout()
plt.show()
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# — assume you’ve already built and sorted `sessions = [(path, mtime), …]` —

fig, ax = plt.subplots(figsize=(6, 6))

# pick one color per session (cycles if >10)
colors = plt.cm.tab10(np.linspace(0, 1, len(sessions)))

for (p, mtime), color in zip(sessions, colors):
    img_file = os.path.join(p, "vis", "com_circle.png")
    if os.path.isfile(img_file):
        img = mpimg.imread(img_file)
        # draw the 0.5 level contour in this color
        ax.contour(img, levels=[0.5], colors=[color], linewidths=1)
    else:
        print(f"No image for {p}")

ax.axis("off")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd

# ─── load your data ───
meta_csv = META
df = pd.read_csv(meta_csv)

# 1) Exact mode (for repeated exact values)
exact_modes = df['r'].mode()
print("Exact mode radius value(s):", exact_modes.values)

# 2) Approximate mode via histogram
hist, edges = np.histogram(df['r'], bins=20)
# find the bin with the highest count
idx = hist.argmax()
# compute its center
mode_bin_center = (edges[idx] + edges[idx+1]) / 2
print(f"Approximate mode (bin center): {mode_bin_center:.2f}")
```
