```python

import numpy as np
import matplotlib.pyplot as plt

# —— 2. 检查范数 —— 
quats=quat
norms = np.linalg.norm(quats, axis=1)
print(f"Quaternion norms: min={norms.min():.4f}, max={norms.max():.4f}, mean={norms.mean():.4f}")

# —— 3. 如有必要，归一化 —— 
tol = 1e-3
if not np.allclose(norms, 1.0, atol=tol):
    quats = quats / norms[:, None]
    print("Detected norms ≠ 1.0 → 已归一化所有四元数")

# —— 4. 计算欧拉角 —— 
rot = R.from_quat(quats)                         # 期待 [x,y,z,w] 格式
angles = rot.as_euler('zyx', degrees=True)       # 返回 [yaw, pitch, roll]
yaw   = angles[:, 0]  # 绕 Z 轴
pitch = angles[:, 1]  # 绕 Y 轴
roll  = angles[:, 2]  # 绕 X 轴

# —— 5. 绘图 —— 
plt.figure(figsize=(10, 6))
plt.plot(yaw,   label='Yaw (Z)')
plt.plot(pitch, label='Pitch (Y)')
plt.plot(roll,  label='Roll (X)')
plt.xlabel('Frame Index')
plt.ylabel('Angle (degrees)')
plt.title('Head Orientation Euler Angles over Time')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# — Next cell: compute & plot head‑direction tuning curve for one ROI —

import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.transform import Rotation as R

# assume `df` is your aligned DataFrame from load_session_data(...)
# and contains columns: qw, qx, qy, qz, time_ms, roi_0, roi_1, …

# 1) quaternion → yaw (degrees 0…360)
quat = df[['qx','qy','qz','qw']].to_numpy()      # SciPy expects [x, y, z, w]
rot = R.from_quat(quat)
yaw = rot.as_euler('zyx', degrees=True)[:, 0] % 360
df['yaw'] = yaw

# 2) bin into e.g. 18 bins (20° each)
n_bins = 18
bin_edges = np.linspace(0, 360, n_bins+1)
bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2
# digitize → 0…n_bins-1
df['dir_bin'] = np.digitize(df['yaw'], bin_edges, right=False) - 1
df['dir_bin'] %= n_bins

# 3) compute mean ΔF/F per bin for ROI 0
roi_idx = 0
tuning = (
    df
    .groupby('dir_bin')[f'roi_{roi_idx}']
    .mean()
    .reindex(range(n_bins))       # ensure all bins present
    .to_numpy()
)

# 4) plot polar tuning curve
theta = np.deg2rad(bin_centers)
fig, ax = plt.subplots(subplot_kw={'projection':'polar'})
ax.plot(theta, tuning)
ax.fill_between(theta, tuning, alpha=0.3)
ax.set_xticks(np.deg2rad(np.arange(0, 360, 45)))
ax.set_title(f'Head‑direction tuning, ROI {roi_idx}')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.transform import Rotation as R

def plot_hd_tuning_three_panels(df, n_bins=18):
    """
    Plot head-direction tuning curves for yaw, pitch, and roll 
    for all ROIs in df, highlighting the single best-tuned ROI in each panel.
    
    df must contain columns:
      - 'qw','qx','qy','qz'
      - 'roi_0', 'roi_1', …, 'roi_N'
    """
    # 1) Convert quaternions → Euler angles (ZYX intrinsic: [yaw, pitch, roll])
    quat = df[['qx','qy','qz','qw']].to_numpy()  # note order [x,y,z,w]
    rot = R.from_quat(quat)
    yaw, pitch, roll = rot.as_euler('zyx', degrees=True).T
    yaw = (yaw + 360) % 360
    roll = (roll + 360) % 360
    df['yaw'], df['pitch'], df['roll'] = yaw, pitch, roll
    
    # 2) Setup subplots
    fig, axes = plt.subplots(3, 1, figsize=(6, 12))
    axes = axes.flatten()
    axes_labels = ['Yaw', 'Pitch', 'Roll']
    angle_data = {'Yaw': df['yaw'], 
                  'Pitch': df['pitch'], 
                  'Roll': df['roll']}
    
    rois = [col for col in df.columns if col.startswith('roi_')]
    
    for ax, axis in zip(axes, axes_labels):
        angles = angle_data[axis]
        
        # 3) Bin edges
        if axis in ('Yaw','Roll'):
            bins = np.linspace(0, 360, n_bins+1)
        else:
            bins = np.linspace(angles.min(), angles.max(), n_bins+1)
        centers = (bins[:-1] + bins[1:]) / 2
        theta = np.deg2rad(centers)
        
        # 4) Compute binned means
        digit = np.digitize(angles, bins) - 1
        digit %= n_bins
        
        tuning = np.zeros((len(rois), n_bins))
        for i, roi in enumerate(rois):
            for b in range(n_bins):
                tuning[i, b] = df.loc[digit == b, roi].mean()
        
        # 5) Compute vector length per ROI
        vec_lengths = np.abs((tuning * np.exp(1j * theta)).sum(axis=1))
        best_idx = np.argmax(vec_lengths)
        
        # 6) Plot all in grey, best in blue
        for i in range(len(rois)):
            ax.plot(theta, tuning[i], color='grey', alpha=0.3)
        ax.plot(theta, tuning[best_idx], color='blue', linewidth=2,
                label=f'Best: {rois[best_idx]}')
        
        ax.set_xticks(np.deg2rad(np.arange(0, 360, 45)))
        ax.set_title(f'Tuning Curves for Head {axis}')
        ax.set_ylabel('Neural Response (ΔF/F)')
        ax.legend(loc='upper right')
    
    axes[-1].set_xlabel('Angle (deg)')
    plt.tight_layout()
    plt.show()

# Usage:
# df = load_session_data(ms_folder)
# plot_hd_tuning_three_panels(df, n_bins=18)
```
