```python
import numpy as np
import pandas as pd

# Dummy COM data: 2 frames, 3 coords, 2 animals
com_data = np.arange(2*3*2).reshape(2, 3, 2)  
# Shape: (frames=2, coords=3, animals=2)
# Values will be easy to spot: 0..11

n_animals = com_data.shape[2]
N = com_data.shape[0]

# xyzxyz style
com_cols_xyzxyz = [f'com{a+1}_{axis}' for a in range(n_animals) for axis in ('x','y','z')]
com_flat_xyzxyz = com_data.reshape(N, 3*n_animals)
df_xyzxyz = pd.DataFrame(com_flat_xyzxyz, columns=com_cols_xyzxyz)

# xxyyzz style
com_cols_xxyyzz = [f'com{a+1}_{axis}' for axis in ('x','y','z') for a in range(n_animals)]
# To get matching data for xxyyzz, reorder before flatten
com_flat_xxyyzz = np.concatenate([com_data[:,axis,:] for axis in range(3)], axis=1)
df_xxyyzz = pd.DataFrame(com_flat_xxyyzz, columns=com_cols_xxyyzz)

print("=== xyzxyz style ===")
print(df_xyzxyz)
print("\n=== xxyyzz style ===")
print(df_xxyyzz)
```
