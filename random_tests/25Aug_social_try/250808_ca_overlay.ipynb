```python
import os, glob
import pandas as pd
import matplotlib.pyplot as plt
from utlis.Ca_tools.roi_spike_vis_utlis import load_minian_data_specific, overlay_all_roi_edges_no_show

def vis_param_simple(session_dir: str,
                     csv_file_path: str,
                     miniscope_subdir: str = "My_V4_Miniscope",
                     display_only: bool = False):
    """Minimal flow: txt → mini_path → selection → pick .nc → overlay → save/show."""
    # --- 0) quick sanity ---
    if not os.path.isdir(session_dir):
        print(f"Session not found: {session_dir}")
        return

    # --- 1) read the only .txt (first nonempty, non-# line) ---
    txts = [f for f in os.listdir(session_dir) if f.endswith(".txt")]
    if len(txts) != 1:
        print(f"Expected exactly 1 txt in {session_dir}, found {len(txts)}")
        return

    mini_base = ""
    with open(os.path.join(session_dir, txts[0]), "r", encoding="utf-8") as f:
        for line in f:
            s = line.strip()
            if s and not s.startswith("#"):
                mini_base = s
                break
    if not mini_base:
        print("No valid mini base path in txt.")
        return

    mini_path = os.path.normpath(os.path.join(mini_base, miniscope_subdir))
    print(f"[mini_path] {mini_path}")

    # --- 2) read CSV (if available) to get selection for this rec_path ---
    selection = None
    if os.path.isfile(csv_file_path):
        try:
            df = pd.read_csv(csv_file_path, usecols=["rec_path", "nc_file"])
            hit = df.loc[df["rec_path"] == mini_path]
            if not hit.empty:
                selection = str(hit.iloc[0]["nc_file"]).strip() or None
        except Exception as e:
            print(f"CSV read warning (ignored): {e}")

    # --- 3) pick an .nc file in miniscope_subdir ---
    nc_file = None
    if selection:
        # try exact/loose match
        matches = glob.glob(os.path.join(mini_path, f"*{selection}*.nc"))
        if matches:
            nc_file = matches[0]

    if nc_file is None:
        # try standard default
        cand = os.path.join(mini_path, "minian_dataset.nc")
        if os.path.exists(cand):
            nc_file = cand

    if nc_file is None:
        # fallback: first .nc in folder
        any_nc = glob.glob(os.path.join(mini_path, "*.nc"))
        if any_nc:
            nc_file = any_nc[0]

    if nc_file is None:
        print("No .nc file found in miniscope subdir.")
        return

    print(f"[nc_file] {nc_file}")

    # --- 4) load + overlay ---
    ts_csv = os.path.join(mini_path, "timeStamps.csv")
    try:
        data, ts = load_minian_data_specific(nc_file, ts_csv)
    except Exception as e:
        print(f"Load data failed: {e}")
        return

    max_proj = data["max_proj"].values
    fig, ax = overlay_all_roi_edges_no_show(data, max_proj)

    # short, stable title
    ax.set_title(os.path.basename(mini_path), fontsize=10)

    # --- 5) output ---
    if display_only:
        plt.show()
    else:
        # Save to miniscope folder
        out_path1 = os.path.join(mini_path, f"overlay_{selection}.png")
        fig.savefig(out_path1, dpi=200, bbox_inches="tight")
        print(f"Saved: {out_path1}")

        # Save to MIR_Aligned folder inside session_dir
        aligned_dir = os.path.join(session_dir, "MIR_Aligned")
        os.makedirs(aligned_dir, exist_ok=True)  # create if missing
        out_path2 = os.path.join(aligned_dir, f"overlay_{selection}.png")
        fig.savefig(out_path2, dpi=200, bbox_inches="tight")
        print(f"Saved: {out_path2}")

    plt.close(fig)
```
