```python
from utlis.exe_engine_utlis.general_table_exe import exec_from_table
from utlis.convert_calib_utlis.chris_wrapper import run_calibration_in_env #batch_calibrate_in_conda_env


from os.path import join, dirname, basename

INTRIN_NAME_fixed = "/data/big_rim/rsync_dcc_sum/2025_09_05/chirs_style_intrinsics"
CONDA_ENV = "sdannce311"

def build_kwargs_for_new_wrapper(r):
    return {
        "test_path": r["rec_path"],
        "intrinsics_dir": r.get("intrinsics_dir") or INTRIN_NAME_fixed,
    }

res = exec_from_table(
    items=filtered_table,
    fn=run_calibration_in_env,
    build=build_kwargs_for_new_wrapper,
    const={
        "conda_env_name": CONDA_ENV,
    },
)

# Check results
for item in res:
    if item['status'] == 'ok':
        result = item['result']
        print(f"{result.get('test_path')}: {result.get('status')}")
```
