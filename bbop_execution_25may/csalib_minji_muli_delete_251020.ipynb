```python
from pathlib import Path

def list_units_with_dupe_label3d_top_only(base_folder: str, suffix="label3d_dannce.mat"):
    base = Path(base_folder)
    for unit in base.glob("*/*"):  # date/rec
        files = [p for p in unit.glob(f"*{suffix}") if p.is_file()]  # top-level only
        if len(files) >= 2:
            print(unit.as_posix())
            for p in files:
                print("  -", p.name)
        else:
            print(f"only found one{files}")

list_units_with_dupe_label3d_top_only("/hpc/group/tdunn/Bryan_Rigs/BigOpenField/minji_social_single_grouphoused")
```
